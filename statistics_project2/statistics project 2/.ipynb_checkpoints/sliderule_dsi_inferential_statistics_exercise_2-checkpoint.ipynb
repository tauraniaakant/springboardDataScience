
### Examining racial discrimination in the US job market

#### Background
Racial discrimination continues to be pervasive in cultures throughout the world. Researchers examined the level of racial discrimination in the United States labor market by randomly assigning identical résumés black-sounding or white-sounding names and observing the impact on requests for interviews from employers.

#### Data
In the dataset provided, each row represents a resume. The 'race' column has two values, 'b' and 'w', indicating black-sounding and white-sounding. The column 'call' has two values, 1 and 0, indicating whether the resume received a call from employers or not.

Note that the 'b' and 'w' values in race are assigned randomly to the resumes.

```python
data = pd.io.stata.read_stata('data/us_job_market_discrimination.dta')
```

# Q1) What test is appropriate for this problem? Does CLT apply?

## Answer: A chi-squared independence test is appropriate for this problem. The CLT applies here

# Q2) What are the null and alternate hypotheses?

## Answer:

    Null Hypothesis: Race does not have a significant impact on the rate of callbacks for resumes. That is, the
    race and callback rate columns are statistically independent.

    Alternate hypothesis: Race does have a significant impact on the rate of callbacks for resumes.

# Q3) Compute margin of error, confidence interval, and p-value.

## Answer:

```python
idata = data[["race", "call"]]
idata = idata.groupby("call").race.value_counts().unstack()
chiSqData = idata
idata = idata.stack().unstack(0)
idata.columns = ["NoCall", "YesCall"]
idata["race_Total"] = idata.NoCall + idata.YesCall
idata = idata.stack().unstack(0)
idata["call_total"] = idata["b"] + idata["w"]
idata
```

```python
chi2, p, ddof, expected = stats.chi2_contingency( chiSqData )
msg = "Test Statistic: {}\np-value: {}\nDegrees of Freedom: {}\n"
print( msg.format( chi2, p, ddof ) )
print( expected )
```

# Q4) Discuss statistical significance.

## Answer:
Since p-value < 0.01, we reject the null hypothesis. Thus race does have significant impact on the callback rate for resumes. The chance of getting a callback rate are statistically significantly higher if a person's name is white-sounding
