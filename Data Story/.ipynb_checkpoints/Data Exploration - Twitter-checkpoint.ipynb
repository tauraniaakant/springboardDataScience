```python
#time series of created overall to get frequency within the day, week, month
#networks to retweets distribution overall and as a timeseries
#
```

```python
tempData = allTweets13Aug.set_index("created_at")
tempData["creation_date"] = tempData.index

minmaxdates = pd.DataFrame(tempData.groupby("user").creation_date.max())
minmaxdates.columns = ['max']
minmaxdates["min"] = tempData.groupby("user").creation_date.min()
minmaxdates["uid"] = range(1,len(minmaxdates.index)+1)

#plt.figure(figsize=(12, 5))

plt.figure(1)
plt.plot(minmaxdates["uid"], minmaxdates["min"], 'r.', linewidth=1.0)
plt.xlabel('User ID',fontsize=14)
plt.ylabel('Earliest year tweet',fontsize=14)
plt.title('Earliest tweet in dataset for each source', fontsize=14)

maxplotDate = minmaxdates["max"].max()+datetime.timedelta(weeks=52)
minplotDate = minmaxdates["min"].min()-datetime.timedelta(weeks=52)
plt.axis([minmaxdates["uid"].min()-1, minmaxdates["uid"].max()+1, minplotDate, maxplotDate])

plt.gca().xaxis.set_ticklabels(minmaxdates.index)
plt.xticks(minmaxdates["uid"])
plt.xticks(rotation=90)
plt.xticks(fontsize=14) 
plt.yticks(fontsize=14)

matplotlib.pyplot.savefig(capstoneDir+'Earliest tweet in dataset for each source-13Aug.svg', bbox_inches='tight')
plt.show()
#py.iplot_mpl(plt.gcf(), stripstyle=True, filename="Earliest tweet in dataset for each source")
```

```python
plt.figure(2)
tempData = allTweets13Aug.set_index("created_at")
tempData["creation_date"] = tempData.index
tempData["day"] = tempData.index.day
dayDF = pd.DataFrame(tempData.groupby("day").retweet_count.count()).reset_index(level="day")
plt.plot(dayDF["day"], dayDF["retweet_count"])
#sns.tsplot(tempData.groupby("day").text.count())
plt.title("Number of tweets by the day of a month",fontsize=14)
plt.xlabel('Day of month',fontsize=14)
plt.ylabel('Number of tweets', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

matplotlib.pyplot.savefig(capstoneDir+'Number of tweets by the day of a month-13Aug.svg', bbox_inches='tight')
plt.show()
#py.iplot_mpl(plt.gcf(), stripstyle=True, filename="Number of tweets by the day of a month")
```

```python
tempData = allTweets13Aug.set_index("created_at")
tempData["creation_date"] = tempData.index
tempData["hour"] = tempData.index.hour
hourDF = pd.DataFrame(tempData.groupby("hour").text.count()).reset_index(level="hour")
plt.plot(hourDF["hour"], hourDF["text"])
#sns.tsplot(tempData.groupby("hour").text.count())
plt.title("Number of tweets by the hour of a day", fontsize=14)
plt.xlabel('Hour of day', fontsize=14)
plt.ylabel('Number of tweets',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

matplotlib.pyplot.savefig(capstoneDir+'Number of tweets by the hour of a day-13Aug.svg', bbox_inches='tight')

plt.show()
#py.iplot_mpl(plt.gcf(), stripstyle=True, filname="Number of tweets by the hour of a day")
```

```python
tempData = allTweets13Aug.set_index("created_at")
tempData["creation_date"] = tempData.index
tempData["minute"] = tempData.index.minute
minuteDF = pd.DataFrame(tempData.groupby("minute").text.count()).reset_index(level="minute")
plt.plot(minuteDF["minute"], minuteDF["text"])
#sns.tsplot(tempData.groupby("minute").text.count())
plt.title("Number of tweets by the minute of an hour",fontsize=14)
plt.xlabel('Minute of hour',fontsize=14)
plt.ylabel('Number of tweets',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

matplotlib.pyplot.savefig(capstoneDir+'Number of tweets by the minute of an hour-13Aug.svg', bbox_inches='tight')
plt.show()
#py.iplot_mpl(plt.gcf(), stripstyle=True, filename="Number of tweets by the minute of an hour")
```

```python
#which time slots in the day get the most retweeted overall, and for different networks
```

```python
tempData = allTweets13Aug.set_index("created_at")
tempData["creation_date"] = tempData.index
tempData["hour"] = tempData.index.hour

hourly_retweetsDF = pd.DataFrame(tempData.groupby("hour").retweet_count.mean())
hourly_retweetsDF = hourly_retweetsDF.reset_index()
hourly_retweetsDF.head()
plt.plot(hourly_retweetsDF["hour"], hourly_retweetsDF["retweet_count"])
plt.title("Number of retweets based on original tweet time",fontsize=14)
plt.xlabel('Original tweet hour of the day',fontsize=14)
plt.ylabel('Number of retweets',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

matplotlib.pyplot.savefig(capstoneDir+'Number of retweets based on original tweet time-13Aug.svg', bbox_inches='tight')
plt.show()
#py.iplot_mpl(plt.gcf(), stripstyle=True, filename="Number of retweets based on original tweet time")
```

```python
tempData = allTweets13Aug.set_index("created_at")
tempData["creation_date"] = tempData.index
tempData["hour"] = tempData.index.hour

hourly_retweetsDF = pd.DataFrame(tempData.groupby(["user","hour"]).retweet_count.mean())
hourly_tweetsDF = pd.DataFrame(tempData.groupby(["user","hour"]).text.count())

fig_count = 0
count = 0

for user in hourly_retweetsDF.unstack().index:
    plt.figure(fig_count+1)
    lineA, = plt.plot(hourly_tweetsDF.loc[hourly_tweetsDF.unstack().index[count]].reset_index()["hour"],
                      hourly_tweetsDF.loc[hourly_tweetsDF.unstack().index[count]])
    lineB, = plt.plot(hourly_retweetsDF.loc[hourly_retweetsDF.unstack().index[count]].reset_index()["hour"],
                      hourly_retweetsDF.loc[hourly_retweetsDF.unstack().index[count]])
    plt.title("Tweets by creation hour for "+str(user), fontsize=14)
    plt.xlabel('Hour of day', fontsize=14)
    plt.ylabel('Number of tweets',fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    #if using matplotlib, then use these for the legend
    lineA.set_label('Tweets by '+str(user))
    lineB.set_label('Re-tweets of '+str(user)+" tweets")
    legend = plt.legend(loc="upper right", bbox_to_anchor=(1.7,1), frameon=True, fancybox=True, framealpha=0.2)
    legend.get_frame().set_facecolor("#00FFCC")

    matplotlib.pyplot.savefig(capstoneDir+"Tweets by creation hour for "+str(user)+"-13Aug.svg", bbox_inches='tight')

    #if using Plotly, then use these instead for the legend
    #py_fig = tls.mpl_to_plotly(plt.gcf(), resize=True)
    #py_fig['layout'].update(showlegend=True, legend=Legend(x=1.05, y=1))
    #py.iplot(py_fig, filename="Tweets by creation hour for "+str(user))
    
    
    fig_count += 1
    count += 1
    
plt.show()
#hourly_retweetsDF.loc[hourly_retweetsDF.unstack().index[0]].reset_index()["hour"]
#hourly_retweetsDF.reset_index(level="hour")
```
