## Inquiry 1: Is there a significant difference in favorites for quotes that are a status and quotes that are not a status?

Null Hypothesis: There is no significant difference between the number of favorites for quotes that are a status and the number of favorites for quotes that are not a status 


Since it is only one variable (quote status) on the effect (favorite count), we do a 2-sample hypothesis test

```python
from scipy import stats
#Variable 1: Quote status
#Group 1: Quote status True
#Group 2: Quote status False
theTweets = allTweets13Aug
theTweets = theTweets[["is_quote_status", "favorite_count"]]

group1 = theTweets[theTweets.is_quote_status == True].favorite_count
group2 = theTweets[theTweets.is_quote_status == False].favorite_count

print(stats.ttest_ind(group1, group2))
print("\nMean favorite_count based on is_quote status:\n"+str(theTweets.groupby("is_quote_status").favorite_count.mean()))
```

Since p-value < 1, there is a significant difference between number of favorites for quoted tweets and the number of tweets that are not quoted tweets. If a tweet is quote, then the favorite count is very low (average of 33 favorites). If a tweet is not a quote, then the favorite count is higher (average of 90 favorites)

## Inquiry 2: Is the above conclusion the same for all news outlets?

```python
theTweets = allTweets13Aug
theTweets = theTweets[["is_quote_status", "retweet_count","user"]]

userList = theTweets.user.unique()

mainDict = dict()

for user in userList:
    tweets = theTweets[theTweets.user == user]
    tweets = tweets.dropna()
    group1 = tweets[tweets.is_quote_status == True].retweet_count
    if len(group1) > 1: #For atleast 1 degree of freedom
        group2 = tweets[tweets.is_quote_status == False].retweet_count
        result = stats.ttest_ind(group1, group2)
        means = tweets.groupby("is_quote_status").retweet_count.mean()
        keys = means.keys()
        subDict = dict()
        subDict["p-value"] = result.pvalue

        for index, value in means.iteritems():

            if str(index) == "False":
                subDict["False count"] = len(group2)
                subDict["Mean if false"] = value
            else:
                subDict["True count"] = len(group1)
                subDict["Mean if true"] = value

        subDict["TotalSampleSize"] = len(group1)+len(group2)
        mainDict[user] = subDict

df = pd.DataFrame.from_dict(mainDict)
df = df.stack().unstack(0)
cols = ["p-value","TotalSampleSize","False count","True count","Mean if false", "Mean if true"]
df = df[cols]

group1 = theTweets[theTweets.is_quote_status == True].retweet_count
group2 = theTweets[theTweets.is_quote_status == False].retweet_count

print(stats.ttest_ind(group1, group2))
df.sort_values("p-value", ascending=False)
```

Interestingly, if the same test is performed on each network, the p-value > 0.05 for all of them. That means, individually, there is not a significant statistical difference in the number of retweets between the tweet being a quote or not. However, overall, as seen earlier, there is a significant statistical difference. A conclusion from this observation is that twitter users retweet fewer times a quote tweet, irrespective of news source.

# Inquiry 4: Is there a relationship between number of retweets or number of favorites and time of original post?

```python
tempData = allTweets13Aug.set_index("created_at")
tempData["creation_date"] = tempData.index
tempData["hour"] = tempData.index.hour

hourly_favoritesDF = pd.DataFrame(tempData.groupby("hour").favorite_count.mean())
hourly_favoritesDF = hourly_favoritesDF.reset_index()
total = hourly_favoritesDF.favorite_count.sum()
hourly_favoritesDF["favorites_percent"] = hourly_favoritesDF.favorite_count/total


hourly_tweetsDF = pd.DataFrame(tempData.groupby("hour").text.count())
hourly_tweetsDF = hourly_tweetsDF.reset_index()
total = hourly_tweetsDF.text.sum()
hourly_tweetsDF["tweets_percent"] = hourly_tweetsDF.text/total

hourly_retweetsDF = pd.DataFrame(tempData.groupby("hour").retweet_count.mean())
hourly_retweetsDF = hourly_retweetsDF.reset_index()
total = hourly_retweetsDF.retweet_count.sum()
hourly_retweetsDF["retweets_percent"] = hourly_retweetsDF.retweet_count/total


plt.plot(hourly_favoritesDF["hour"], hourly_favoritesDF["favorites_percent"])
plt.plot(hourly_tweetsDF["hour"], hourly_tweetsDF["tweets_percent"])
plt.plot(hourly_retweetsDF["hour"], hourly_retweetsDF["retweets_percent"])
fig = plt.gcf()
fig.set_size_inches(10, 5)
plt.title("Number of favorites based on original tweet time",fontsize=14)
plt.xlabel('Original tweet hour of the day',fontsize=14)
plt.ylabel('Number of favorites',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
legend = plt.legend(loc="upper right", bbox_to_anchor=(1,1), frameon=True, fancybox=True, framealpha=0.2, fontsize=14)
legend.get_frame().set_facecolor("#00FFCC")
plt.show()
#py.iplot_mpl(plt.gcf(), stripstyle=True, filename="Number of favorites based on original tweet time")
```

From this plot, we can see that, both the the percent of retweets and favorites spike at around 2-5am and dips throughout the day. However, the number of tweets being published is the lowest during early morning and picks up and peaks in the evening. We will now conduct a test of significance.

Tests: 

Chi-squared independence test 1: Between 2 variables: hour of day and whether it was retweeted

Chi-squared independence test 2: Between 2 variables: hour of day and whether it was retweeted

```python
df = allTweets13Aug[["created_at","favorite_count", "retweet_count"]]

df['hour'] = df.created_at.dt.hour
df = df.drop("created_at",axis=1)
#retweetDf = df.groupby("hour").retweet_count.value_counts().sort_index()
#retweetDf = retweetDf.unstack(0)
#favoriteDf = df.groupby("hour").favorite_count.value_counts().sort_index()
#stats.chisquare(retweetDf)

hours = np.arange(0,24,1)

p_values = dict()

#for hour in hours:

retweetDf = df[["hour", "retweet_count"]]
retweetDf["retweeted"] = retweetDf.retweet_count > 0
retweetDf = retweetDf.reset_index()
retweetDf["phase"] = np.where(retweetDf.hour >= 10, "10amOnwards", "Before10am")
retweetDf = retweetDf.drop("hour", axis=1)
retweetDf = retweetDf.groupby("phase")['retweeted'].value_counts()
retweetDf = retweetDf.unstack(0)
print(stats.chi2_contingency(retweetDf))
print(retweetDf)
retweetDf = df[["hour", "retweet_count"]]
retweetDf.head()
retweetDf["phase"] = np.where(retweetDf.hour >= 10, "10amOnwards", "Before10am")
retweetDf = retweetDf.drop("hour", axis=1)
retweetDf = retweetDf.groupby("phase")['retweet_count'].mean()
print(pd.DataFrame(retweetDf))
print("Posts after 10 AM receive "+str(abs(retweetDf.iloc[0]-retweetDf.iloc[1])/(retweetDf.iloc[0])*100)+
      "% less retweets than those posted before 10 AM")
```

Since p << 0.01, there is statistically signigicant evidence that there is a relationship between a posts that were posted before 10am and posts after 10am. For posts before 10am, the number of retweets posts receive is higher than posts after 10am.
